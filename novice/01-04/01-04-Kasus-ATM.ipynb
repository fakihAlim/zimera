### Kasus ATM

```python
class Deposit(ATMMachine):
    def __init__(self):
        self.deposit = 0

    def setDeposit(self, d):
        self.deposit = d

    def getDeposit(self):
        return self.deposit


class Withdraw:
    withdraw = 0

    def setWithdraw(self, w):
        self.withdraw = w

    def getWithdraw(self):
        return self.withdraw


class BalanceInquiry:
    balance = 0

    def __init__(self):
        self.balance = BalanceInquiry.balance

    def setBalance(self, b):
        self.balance = b

    def getBalance(self):
        return self.balance


class ATMMachine:
    def __init__(self):
        self.BalanceInquiry = BalanceInquiry
        self.Deposit = Deposit
        self.Withdraw = Withdraw

    def check_balance(self):
        print("\tYour current balance is ", self.BalanceInquiry.balance)

    def withdrawMoney(self):
        keluarDuit = Withdraw()

        if self.BalanceInquiry.balance == 0:
            print("\tYour current balance is zero.")
        elif self.BalanceInquiry.balance <= 500:
            print("\tYou do not have sufficient money to withdraw")
        elif self.Withdraw.withdraw > self.BalanceInquiry.balance:
            print("\tThe amount you withdraw is greater than to your balance")
        else:
            self.BalanceInquiry.balance -= self.Withdraw.withdraw
            getuangnya = keluarDuit.getWithdraw()
            print("\n\tYou withdraw the amount of Php ", getuangnya)

    def depositMoney(self):
        # jika ingin memanggil methodnya tidak bisa langsung harus dibuatkan objeknya
        bi = BalanceInquiry()
        uangnya = bi.getBalance()
        print("\tYou deposited the amount of ", uangnya)


select = 0

while True:
    print("Welcome to this simple ATM Machine")
    print("Please select ATM transaction!")
    print("[1] Deposit")
    print("[2] Withdraw")
    print("[3] Balance Inquiry")
    print("[4] Exit")

    select = int(input("What would you like to do? "))
    atm = ATMMachine()
    atm.check_balance()

    if select == 1:
        atm.Deposit.deposit = int(input("Enter the amount of money to deposit: "))
        atm.BalanceInquiry.balance = atm.Deposit.deposit + atm.BalanceInquiry.balance
        atm.depositMoney()
    elif select == 2:
        print(
            "\n\tTo withdraw, make sure that you have sufficient balance in your account."
        )
        atm.Withdraw.withdraw = int(input("Enter amount of money to withdraw: "))
        atm.withdrawMoney()
    elif select == 3:
        atm.check_balance()
    else:
        print("\n\tTransaction exited.")
        break

print("\n\tThank you for using this simple ATM Machine.")
```
